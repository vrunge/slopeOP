## 1. Data generation

Here we generate four signals, trough four different functions :

- signal_1()
- signal_2()
- signal_3()
- signal_4()

Each function returns a specific signal made with 500 points.

You can add a gaussian noise whose mean equals zero and sigma equals the value you pass trough the "noise" argument. You can also fix the seed through the "seed" argument.

#### The four plotted signals, with no noise

```python
fig, ax = plt.subplots(2,2)

ax[0, 0].plot(signal_1())
ax[0, 1].plot(signal_2())
ax[1, 0].plot(signal_3())
ax[1, 1].plot(signal_4());
```

Now we know how to generate our four signals, let's generate them with different noise and seed. We will generate them with four different noises (sigma) : 0 (no noise), 3, 12 and 24. For each noise, we will generate 100 different versions by changing the seed, from 1 to 100.

So at the end we will have 301 versions of each signal : three sigma * hundred seeds + the original signal.

Let's save each batch in four different files, one file per signal.

```python
# Variable initialization
seeds = np.arange(1,11)
sigmas = np.array([3,12,24])

# Loop over each signal
for f in [signal_1, signal_2, signal_3, signal_4,signal_1_at_50pt, signal_1_at_300pt]:
    
    signal={}

    column_format = "{sigma}_{seed}"
    column_name = column_format.format(sigma=0, seed=0)
    signal[column_name] = f(noise=0)
    
    for sigma in sigmas:
        for seed in seeds:
            column_name = column_format.format(sigma=sigma, seed=seed)
            signal[column_name] = f(noise=sigma,seed=seed)

    # Write data
    file_name = f"{f.__name__}.csv"
    pd.DataFrame(signal).to_csv(file_name, index=False)
```

Each file is a CSV dedicated to one specific signal. Each column represents a different version and each row represents a specific data point (from 1 to 500). So the dimension of each CSV is (n * v) where n = 500 (the number of datapoints) and v = the number of different versions, which is 301 (the original signal + (3 noises * 100 seeds)).

```python
def get_mse_vs_beta(signal_name, algo, betas, sigma, seeds=np.arange(1,11)):
    
    mses = np.zeros(len(betas))
    for i,beta in enumerate(betas):
        signal_data = pd.read_csv(f"simu_{signal_name}_{algo}_{beta:.3f}.csv")
        mse_temp = []
        for seed in seeds:
            mse_temp.append(mse(signal_data["X0_0"], signal_data[f"X{sigma}_{seed}"]))
        mses[i] = np.mean(mse_temp)
        
    return np.log(mses)

def get_rmses(signal_name, algo, beta, sigma, seeds=np.arange(1,11)):
    
    mses = []
    signal_data = pd.read_csv(f"simu_{signal_name}_{algo}_{beta:.3f}.csv")
    for seed in seeds:
        mses.append(mse(signal_data["X0_0"], signal_data[f"X{sigma}_{seed}"]))
        
    return np.sqrt(mses)

marker_algo = {
    'Slope OP': '.',
    'CPOP': 'o',
    'op2d': '*',
    'fPop': 'o',
    'RfPop': '*'
}

fig, ax = plt.subplots(4,3, figsize=(40,26))

# First row
signal = "signal_1_at_50pt"
original_signal = pd.read_csv(f"{signal}.csv")
ax[0,0].plot(original_signal["12_1"], '.')
ax[0,0].plot(original_signal["0_0"])

signal = "signal_1_at_300pt"
original_signal = pd.read_csv(f"{signal}.csv")
ax[0,1].plot(original_signal["12_1"], '.')
ax[0,1].plot(original_signal["0_0"])

signal = "signal_1"
original_signal = pd.read_csv(f"{signal}.csv")
ax[0,2].plot(original_signal["12_1"], '.')
ax[0,2].plot(original_signal["0_0"])

# Second row
signals = ["signal_1_at_50pt", "signal_1_at_300pt", "signal_1"]
betas = np.arange(0.1,5.1,0.1)

for i, signal in enumerate(signals):
    original_signal = pd.read_csv(f"{signal}.csv")
    datas = {
        'Slope OP': get_mse_vs_beta(signal, "slopeOP", betas, 12),
        'fPop': get_mse_vs_beta(signal, "fpop", betas, 12),
        'RfPop': get_mse_vs_beta(signal, "rfpop", betas, 12)
    }
    
    for algo in datas:
        ax[1,i].plot(betas, datas[algo], linewidth=0, marker=marker_algo[algo], label=algo)    
    ax[1,i].legend()
    
# Third row
signals = ["signal_1_at_50pt", "signal_1_at_300pt", "signal_1"]
algo = "fpop"
beta = 0.1

for i, signal in enumerate(signals):
    signal = signal
    original_signal = pd.read_csv(f"{signal}.csv")["12_1"]
    fpop_signal = pd.read_csv(f"simu_{signal}_{algo}_{beta:.3f}.csv")
    ax[2,i].plot(original_signal, '.')
    ax[2,i].plot(fpop_signal["X12_1"])
    
# Fourth row
signals = ["signal_1_at_50pt", "signal_1_at_300pt", "signal_1"]
algo = "fpop"
beta = 1.8

for i, signal in enumerate(signals):
    signal = signal
    original_signal = pd.read_csv(f"{signal}.csv")["12_1"]
    fpop_signal = pd.read_csv(f"simu_{signal}_{algo}_{beta:.3f}.csv")
    ax[3,i].plot(original_signal, '.')
    ax[3,i].plot(fpop_signal["X12_1"])
```
